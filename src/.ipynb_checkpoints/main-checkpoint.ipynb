# Definições globais

```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras import Input, Model
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Conv2DTranspose, Dropout, concatenate, Add, Activation, Multiply, UpSampling2D
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical
import tensorflow.keras.backend as K
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
import pandas as pd
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')

# Definição de parâmetros gerais
IMAGE_SIZE = 512
BATCH_SIZE = 4
EPOCHS = 100
```

## Modelo U-Net

```python
# Função para construir o modelo U-Net
def build_unet_model(input_shape=(IMAGE_SIZE, IMAGE_SIZE, 3), n_classes=1, activation='sigmoid'):
    
    inputs = Input(input_shape)
    s = inputs

    c1 = Conv2D(16, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same')(s)
    c1 = Dropout(0.1)(c1)
    c1 = Conv2D(16, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same')(c1)
    p1 = MaxPooling2D((2, 2))(c1)

    c2 = Conv2D(32, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same')(p1)
    c2 = Dropout(0.1)(c2)
    c2 = Conv2D(32, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same')(c2)
    p2 = MaxPooling2D((2, 2))(c2)

    c3 = Conv2D(64, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same')(p2)
    c3 = Dropout(0.2)(c3)
    c3 = Conv2D(64, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same')(c3)
    p3 = MaxPooling2D((2, 2))(c3)

    c4 = Conv2D(128, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same')(p3)
    c4 = Dropout(0.2)(c4)
    c4 = Conv2D(128, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same')(c4)
    p4 = MaxPooling2D(pool_size=(2, 2))(c4)

    c5 = Conv2D(256, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same')(p4)
    c5 = Dropout(0.3)(c5)
    c5 = Conv2D(256, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same')(c5)

    u6 = Conv2DTranspose(128, (2, 2), strides=(2, 2), padding='same')(c5)
    u6 = concatenate([u6, c4])
    c6 = Conv2D(128, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same')(u6)
    c6 = Dropout(0.2)(c6)
    c6 = Conv2D(128, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same')(c6)

    u7 = Conv2DTranspose(64, (2, 2), strides=(2, 2), padding='same')(c6)
    u7 = concatenate([u7, c3])
    c7 = Conv2D(64, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same')(u7)
    c7 = Dropout(0.2)(c7)
    c7 = Conv2D(64, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same')(c7)

    u8 = Conv2DTranspose(32, (2, 2), strides=(2, 2), padding='same')(c7)
    u8 = concatenate([u8, c2])
    c8 = Conv2D(32, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same')(u8)
    c8 = Dropout(0.1)(c8)
    c8 = Conv2D(32, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same')(c8)

    u9 = Conv2DTranspose(16, (2, 2), strides=(2, 2), padding='same')(c8)
    u9 = concatenate([u9, c1], axis=3)
    c9 = Conv2D(16, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same')(u9)
    c9 = Dropout(0.1)(c9)
    c9 = Conv2D(16, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same')(c9)

    outputs = Conv2D(n_classes, (1, 1), activation=activation)(c9)
    model = Model(inputs=[inputs], outputs=[outputs])
    return model
```

## Avaliação

```python
def iou(y_true, y_pred, n_classes=1):
    y_true = K.flatten(y_true)
    y_pred = K.flatten(y_pred)
    
    intersection = K.sum(K.cast(y_true * y_pred, 'float32'))
    union = K.sum(K.cast(y_true + y_pred, 'float32')) - intersection
    
    iou_value = (intersection + K.epsilon()) / (union + K.epsilon())
    
    if n_classes == 1:
        return iou_value
    else:
        # No caso de multi-classe, itera sobre cada classe e calcula o IoU médio
        iou_per_class = []
        for i in range(n_classes):
            y_true_class = K.cast(K.equal(y_true, i), 'float32')
            y_pred_class = K.cast(K.equal(y_pred, i), 'float32')
            
            intersection_class = K.sum(K.cast(y_true_class * y_pred_class, 'float32'))
            union_class = K.sum(K.cast(y_true_class + y_pred_class, 'float32')) - intersection_class
            
            iou_class = (intersection_class + K.epsilon()) / (union_class + K.epsilon())
            iou_per_class.append(iou_class)
        
        # Retorna a média do IoU por classe
        return K.mean(K.stack(iou_per_class))


# Função para avaliar o modelo e calcular métricas, incluindo AUC
def evaluate_model(model, test_dataset, model_name, n_classes=1):
    y_true = []
    y_pred = []

    for img, mask in test_dataset:
        preds = model.predict(img)
        if n_classes == 1:
            preds = (preds > 0.5).astype("int32")
            y_true.append(mask.numpy().astype("int32"))  # Mantém o mask como tensor
        else:
            preds = tf.argmax(preds, axis=-1)  # Seleciona a classe com maior probabilidade
            mask = tf.argmax(mask, axis=-1)
            y_true.append(mask.numpy())  # Converte para numpy para avaliação

        # Ajuste aqui: Usar tf.reshape() para achatar o tensor em vez de flatten()
        y_pred.append(tf.reshape(preds, [-1]).numpy())  # Achata o tensor e converte para numpy

    # Concatenar resultados
    y_true = np.concatenate([np.reshape(mask, [-1]) for mask in y_true])  # Achatar y_true também
    y_pred = np.concatenate(y_pred)

    # Calcular métricas
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred, average='weighted', zero_division=1)
    recall = recall_score(y_true, y_pred, average='weighted')
    f1 = f1_score(y_true, y_pred, average='weighted')
    iou_score = iou(y_true, y_pred, n_classes)
    
    # Calcular AUC Score (para segmentação binária)
    if n_classes == 1:
        auc_score = roc_auc_score(y_true, y_pred)
    else:
        auc_score = roc_auc_score(y_true, y_pred, average='weighted', multi_class='ovo')

    # Exibir métricas
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1 Score: {f1:.4f}")
    print(f"IoU Score: {iou_score:.4f}")
    print(f"AUC Score: {auc_score:.4f}")
    
    # Retornar as métricas como dicionário
    return {
        'model': model_name,  # Nome do modelo
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1_score': f1,
        'iou_score': iou_score,
        'auc_score': auc_score  # Incluindo AUC no retorno
    }
```

```python
# Função para plotar gráficos de perda e acurácia
def plot_metrics(history, title):
    epochs = range(1, len(history['loss']) + 1)
    plt.figure(figsize=(12, 4))
    
    # Gráfico de perda
    plt.subplot(1, 2, 1)
    plt.plot(epochs, history['loss'], 'b-', label='Loss')
    plt.plot(epochs, history['val_loss'], 'r-', label='Val Loss')
    plt.title(f'{title} - Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()

    # Gráfico de acurácia
    plt.subplot(1, 2, 2)
    plt.plot(epochs, history['accuracy'], 'b-', label='Accuracy')
    plt.plot(epochs, history['val_accuracy'], 'r-', label='Val Accuracy')
    plt.title(f'{title} - Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    
    plt.show()
```

## Dataset

```python
# Definição dos diretórios de dados
root = '../data/raw'
input_train_data = os.path.join(root, 'training/input')
target_train_data = os.path.join(root, 'training/target')
input_test_data = os.path.join(root, 'test/input')
target_test_data = os.path.join(root, 'test/target')
exts = ('png',)  # Considerando apenas imagens .png

# Função para ler os arquivos de imagem e máscara
def read_files(image_path, mask=False):
    image = tf.io.read_file(image_path)
    if mask:
        image = tf.image.decode_png(image, channels=1)  # Para imagens em escala de cinza
        image = tf.image.resize(images=image, size=[IMAGE_SIZE, IMAGE_SIZE])
        image = tf.cast(image, tf.float32) / 255.0  # Normaliza a máscara
    else:
        image = tf.image.decode_png(image, channels=3)  # Para imagens RGB
        image = tf.image.resize(images=image, size=[IMAGE_SIZE, IMAGE_SIZE])
        image = tf.cast(image, tf.float32) / 255.0  # Normaliza a imagem
    return image

# Função para carregar os dados
def load_data(image_list, mask_list, n_classes=1):
    image = read_files(image_list)
    mask = read_files(mask_list, mask=True)
    if n_classes > 1:
        mask = tf.squeeze(mask, axis=-1)
        mask = to_categorical(mask, num_classes=n_classes)  # Conversão para one-hot encoding
    return image, mask

# Função geradora de dados ajustada para aceitar n_classes opcionalmente
def data_generator(image_list, mask_list, n_classes=1):
    dataset = tf.data.Dataset.from_tensor_slices((image_list, mask_list))
    dataset = dataset.map(lambda x, y: load_data(x, y, n_classes), num_parallel_calls=tf.data.AUTOTUNE)
    dataset = dataset.batch(BATCH_SIZE, drop_remainder=False)
    return dataset

# Preparação dos caminhos dos arquivos de treinamento e teste
train_images = sorted(
    [
        os.path.join(input_train_data, fname)
        for fname in os.listdir(input_train_data)
        if fname.endswith(exts) and not fname.startswith(".")
    ]
)

train_masks = sorted(
    [
        os.path.join(target_train_data, fname)
        for fname in os.listdir(target_train_data)
        if fname.endswith(exts) and not fname.startswith(".")
    ]
)

test_images = sorted(
    [
        os.path.join(input_test_data, fname)
        for fname in os.listdir(input_test_data)
        if fname.endswith(exts) and not fname.startswith(".")
    ]
)

test_masks = sorted(
    [
        os.path.join(target_test_data, fname)
        for fname in os.listdir(target_test_data)
        if fname.endswith(exts) and not fname.startswith(".")
    ]
)
```

# Aplicação de U-Net básica

A seguir será treinado um modelo do tipo U-Net e avaliado seu desempenho no dataset DRIVE.

```python
# Geração dos datasets de treinamento e teste
train_dataset = data_generator(train_images, train_masks)
test_dataset = data_generator(test_images, test_masks)
```

```python
# Treinamento com Sigmoid
print("Treinamento com Sigmoid")
model_sigmoid = build_unet_model(activation='sigmoid')
model_sigmoid.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

history_sigmoid = model_sigmoid.fit(train_dataset, epochs=EPOCHS, validation_data=test_dataset, verbose=2)

# Avaliação do modelo com Sigmoid
print("Avaliação com Sigmoid:")
metrics = evaluate_model(model_sigmoid, test_dataset, 'sigmoid_model', n_classes=1)
model_metrics.append(metrics)
```

## Balanceamento vs Desbalanceamento

```python
# Função para calcular pesos de classes
def calculate_class_weights(mask_list, n_classes):
    # Contagem das classes em todas as máscaras
    total_counts = np.zeros(n_classes)
    for mask_path in mask_list:
        mask = tf.io.read_file(mask_path)
        mask = tf.image.decode_png(mask, channels=1)
        mask = np.array(mask).flatten()
        
        # Filtrar valores que não correspondem às classes esperadas
        mask = mask[mask < n_classes]
        
        unique, counts = np.unique(mask, return_counts=True)
        total_counts[unique] += counts
    
    # Cálculo do inverso da frequência relativa para cada classe
    total_counts = np.maximum(total_counts, 1)  # Para evitar divisão por zero
    weights = 1.0 / total_counts
    weights /= np.sum(weights)  # Normaliza para que a soma seja 1
    return {i: weights[i] for i in range(n_classes)}

# Calcular os pesos das classes
class_weights = calculate_class_weights(train_masks, n_classes=2)


print("Treinamento com Sigmoid (Com tratamento de desbalanceamento - Class Weights)")
model_sigmoid_weights = build_unet_model(n_classes=1, activation='sigmoid')
model_sigmoid_weights.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Treinar o modelo com tratamento de desbalanceamento
history_weights = model_sigmoid_weights.fit(train_dataset, epochs=EPOCHS, class_weight=class_weights, validation_data=test_dataset, verbose=2)

# Avaliação com tratamento de desbalanceamento
print("Avaliação com Sigmoid (Com tratamento de desbalanceamento - Class Weights):")
metrics = evaluate_model(model_sigmoid_weights, test_dataset,
                         'sigmoid_with balancing', n_classes=1)
model_metrics.append(metrics)
plot_metrics(history_weights.history, 'Sigmoid com Balanceamento)')
plot_metrics(history_sigmoid.history, 'Sigmoid sem Balanceamento')
```

Comparação das Métricas de Desempenho:
    
1. Função de Perda (Loss)

- Modelo com Balanceamento:

A curva de perda para o modelo treinado com dados balanceados apresenta um comportamento anômalo. A perda de treinamento (linha azul) é consistentemente próxima de zero, sugerindo que o modelo está ajustado de forma extrema aos dados de treinamento. Já a perda de validação (linha vermelha) é bastante alta e flutua muito, sem uma tendência clara de convergência, permanecendo acima de 1.5 ao longo de todas as épocas.
Esse comportamento sugere que o balanceamento aplicado pode ter causado um ajuste excessivo do modelo aos dados de treinamento, mas que ele não está conseguindo generalizar para os dados de validação. O desbalanceamento natural dos dados pode ser significativo e, neste caso, a estratégia de balanceamento parece ter comprometido o processo de aprendizado.

- Modelo sem Balanceamento:

No caso do modelo treinado sem balanceamento, a curva de perda exibe um comportamento muito mais estável. A perda de treinamento começa alta (~0.5), mas diminui gradualmente e se estabiliza em torno de 0.1 após cerca de 50 épocas. A perda de validação segue de perto a curva de treinamento, também estabilizando-se por volta de 0.1.
Esse comportamento indica que o modelo está conseguindo aprender os padrões dos dados de forma estável e generalizando bem para o conjunto de validação, sem grandes oscilações.

2. Acurácia (Accuracy)

- Modelo com Balanceamento:

A acurácia do modelo com balanceamento começa alta (~0.20) nas primeiras épocas, mas cai drasticamente logo em seguida, convergindo para valores em torno de 0.05 e 0.06. Esse comportamento reflete o baixo desempenho do modelo em generalizar para os dados de validação, com flutuações significativas ao longo do treinamento. Mesmo após 100 épocas, a acurácia de validação permanece em níveis extremamente baixos, indicando que o modelo não está aprendendo adequadamente.
A acurácia de treinamento é igualmente baixa e segue o mesmo padrão da validação, sugerindo que o modelo está sendo incapaz de extrair informações relevantes dos dados.

- Modelo sem Balanceamento:

O modelo sem balanceamento, por outro lado, começa com uma acurácia inicial mais baixa (~0.75), mas melhora consistentemente ao longo das épocas, alcançando aproximadamente 0.92 no final do treinamento. A curva de validação acompanha a curva de treinamento de perto, com diferenças mínimas entre as duas, o que demonstra que o modelo está generalizando bem.
A ausência de flutuações abruptas e o comportamento estável da acurácia sugerem que o modelo sem balanceamento foi capaz de lidar melhor com a natureza desbalanceada dos dados.

3. Conclusão da Análise:

Com base nos gráficos apresentados, é evidente que o modelo treinado sem balanceamento apresentou um desempenho significativamente melhor em comparação com o modelo que utilizou uma estratégia de balanceamento. O modelo com balanceamento sofreu com uma perda de validação alta e instável, e sua acurácia tanto de treinamento quanto de validação foi muito baixa ao longo de todas as épocas. Esses sinais indicam que o balanceamento não foi eficaz para este conjunto de dados e, ao contrário, parece ter prejudicado o aprendizado do modelo.

Já o modelo sem balanceamento demonstrou uma perda consistente e uma acurácia robusta, tanto no treinamento quanto na validação, sem sinais de overfitting ou underfitting significativos. As curvas estáveis e a alta acurácia final (~0.92) indicam que o modelo conseguiu aprender bem os padrões presentes no conjunto de dados, mesmo sem uma intervenção para balancear as classes.

Com base nesses resultados, foi decidido não utilizar uma estratégia de balanceamento para as próximas etapas do projeto. Embora o desbalanceamento das classes seja uma preocupação válida, os experimentos mostraram que o balanceamento aplicado resultou em uma degradação significativa no desempenho do modelo. O modelo treinado sem balanceamento demonstrou melhor desempenho, tanto em termos de perda quanto de acurácia, e se mostrou mais robusto em termos de generalização.

Portanto, a opção por não utilizar balanceamento foi tomada com o intuito de preservar a capacidade do modelo de generalizar e aprender de maneira eficaz com o conjunto de dados original.

## U-Net com Backbone (ResNet50)

```python
# Treinamento do modelo ResNet50 com Fine-tuning

def build_unet_with_backbone(input_shape=(512, 512, 3), n_classes=1):
    backbone = ResNet50(include_top=False, weights='imagenet', input_shape=input_shape)
    backbone.trainable = False  # Congela os pesos do backbone para evitar sobreajuste

    # Encoder
    inputs = backbone.input
    c1 = backbone.get_layer("conv1_relu").output  # 256x256
    c2 = backbone.get_layer("conv2_block3_out").output  # 128x128
    c3 = backbone.get_layer("conv3_block4_out").output  # 64x64
    c4 = backbone.get_layer("conv4_block6_out").output  # 32x32
    c5 = backbone.get_layer("conv5_block3_out").output  # 16x16

    # Decoder com camadas transpostas para garantir a mesma dimensão
    u6 = UpSampling2D(size=(2, 2))(c5)  # 16x16 -> 32x32
    u6 = concatenate([u6, c4])
    c6 = Conv2D(512, (3, 3), activation='relu', padding='same')(u6)

    u7 = UpSampling2D(size=(2, 2))(c6)  # 32x32 -> 64x64
    u7 = concatenate([u7, c3])
    c7 = Conv2D(256, (3, 3), activation='relu', padding='same')(u7)

    u8 = UpSampling2D(size=(2, 2))(c7)  # 64x64 -> 128x128
    u8 = concatenate([u8, c2])
    c8 = Conv2D(128, (3, 3), activation='relu', padding='same')(u8)

    u9 = UpSampling2D(size=(2, 2))(c8)  # 128x128 -> 256x256
    u9 = concatenate([u9, c1])
    c9 = Conv2D(64, (3, 3), activation='relu', padding='same')(u9)

    # Última camada de upsampling para 512x512 (sem ultrapassar)
    u10 = UpSampling2D(size=(2, 2))(c9)  # 256x256 -> 512x512

    # Saída final com a dimensão correta
    outputs = Conv2D(n_classes, (1, 1), activation='sigmoid')(u10)  # Mantendo ativação sigmoid para segmentação binária
    model = Model(inputs=[inputs], outputs=[outputs])
    return model



# Iniciando o fine-tuning após algumas épocas de treinamento
model_backbone = build_unet_with_backbone()
model_backbone.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Treinamento com a ResNet50 congelada nas primeiras 20 épocas
history_backbone = model_backbone.fit(train_dataset, epochs=20, validation_data=test_dataset, verbose=2)

# Agora descongelar as últimas camadas para fine-tuning
for layer in model_backbone.layers[-20:]:
    layer.trainable = True

# Treinamento com camadas descongeladas
history_backbone_finetuned = model_backbone.fit(train_dataset, epochs=EPOCHS, validation_data=test_dataset, verbose=2)

# Avaliação do modelo com backbone descongelado
metrics = evaluate_model(model_backbone, test_dataset, 'sigmoid_with_backbone')


model_metrics.append(metrics)

# Plotar métricas para U-Net com Backbone
plot_metrics(history_backbone.history, 'U-Net com Backbone ResNet50')
```

A perda de validação (linha vermelha) segue um comportamento similar, sugerindo que o modelo está generalizando bem para os dados de validação. A convergência precoce de ambas as curvas indica que o modelo ajustou-se de forma eficiente, aprendendo rapidamente os padrões relevantes do conjunto de dados.

A partir da 5ª época, a acurácia de treinamento (linha azul) e de validação (linha vermelha) estabilizam-se em torno de 0.88-0.89, com curvas praticamente idênticas.

## U-Net com Camadas de Atenção

```python
# Função de Atenção
def attention_block(x, g, inter_channel):
    theta_x = Conv2D(inter_channel, (1, 1))(x)
    phi_g = Conv2D(inter_channel, (1, 1))(g)
    add = Add()([theta_x, phi_g])
    act = Activation('relu')(add)  # Usar Activation layer em vez de tf.nn.relu
    psi = Conv2D(1, (1, 1), activation='sigmoid')(act)
    mul = Multiply()([x, psi])
    return mul

# U-Net com Attention
def build_attention_unet(input_shape=(IMAGE_SIZE, IMAGE_SIZE, 3), n_classes=1):
    inputs = Input(input_shape)
    s = inputs

    c1 = Conv2D(16, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same')(s)
    c1 = Dropout(0.1)(c1)
    c1 = Conv2D(16, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same')(c1)
    p1 = MaxPooling2D((2, 2))(c1)

    c2 = Conv2D(32, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same')(p1)
    c2 = Dropout(0.1)(c2)
    c2 = Conv2D(32, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same')(c2)
    p2 = MaxPooling2D((2, 2))(c2)

    c3 = Conv2D(64, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same')(p2)
    c3 = Dropout(0.2)(c3)
    c3 = Conv2D(64, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same')(c3)
    p3 = MaxPooling2D((2, 2))(c3)

    c4 = Conv2D(128, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same')(p3)
    c4 = Dropout(0.2)(c4)
    c4 = Conv2D(128, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same')(c4)
    p4 = MaxPooling2D(pool_size=(2, 2))(c4)

    c5 = Conv2D(256, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same')(p4)
    c5 = Dropout(0.3)(c5)
    c5 = Conv2D(256, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same')(c5)

    # Atenção e Decodificação
    u6 = Conv2DTranspose(128, (2, 2), strides=(2, 2), padding='same')(c5)
    u6 = attention_block(u6, c4, 128)
    u6 = concatenate([u6, c4])
    c6 = Conv2D(128, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same')(u6)
    c6 = Dropout(0.2)(c6)
    c6 = Conv2D(128, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same')(c6)

    u7 = Conv2DTranspose(64, (2, 2), strides=(2, 2), padding='same')(c6)
    u7 = attention_block(u7, c3, 64)
    u7 = concatenate([u7, c3])
    c7 = Conv2D(64, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same')(u7)
    c7 = Dropout(0.2)(c7)
    c7 = Conv2D(64, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same')(c7)

    u8 = Conv2DTranspose(32, (2, 2), strides=(2, 2), padding='same')(c7)
    u8 = attention_block(u8, c2, 32)
    u8 = concatenate([u8, c2])
    c8 = Conv2D(32, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same')(u8)
    c8 = Dropout(0.1)(c8)
    c8 = Conv2D(32, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same')(c8)

    u9 = Conv2DTranspose(16, (2, 2), strides=(2, 2), padding='same')(c8)
    u9 = attention_block(u9, c1, 16)
    u9 = concatenate([u9, c1], axis=3)
    c9 = Conv2D(16, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same')(u9)
    c9 = Dropout(0.1)(c9)
    c9 = Conv2D(16, (3, 3), activation='elu', kernel_initializer='he_normal', padding='same')(c9)

    outputs = Conv2D(n_classes, (1, 1), activation='sigmoid')(c9)
    model = Model(inputs=[inputs], outputs=[outputs])
    return model


# Treinar e avaliar o modelo Attention U-Net
print("Treinamento com Attention U-Net:")
model_attention = build_attention_unet()
model_attention.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
history_attention = model_attention.fit(train_dataset, epochs=EPOCHS, validation_data=test_dataset, verbose=2)
metrics = evaluate_model(model_attention, test_dataset, 'sigmoid_with_attention')
model_metrics.append(metrics)

# Plotar métricas para Attention U-Net
plot_metrics(history_attention.history, 'Attention U-Net')
```

1. Função de Perda (Loss)

O gráfico de perda (à esquerda) mostra que o modelo Attention U-Net começa com uma perda relativamente alta, acima de 0.45, mas a função de perda diminui rapidamente nas primeiras 10 épocas e continua a convergir de forma mais gradual ao longo do treinamento.
Há algumas oscilações notáveis na perda de validação (linha vermelha), especialmente por volta da 40ª e da 80ª épocas, o que indica algumas dificuldades temporárias na generalização do modelo nessas fases. No entanto, essas oscilações não são muito pronunciadas e a perda acaba se estabilizando em torno de 0.12-0.15.
A perda de treinamento (linha azul) segue de perto a curva de validação, o que sugere que o modelo não sofre de overfitting severo, mas há uma leve diferença durante algumas épocas que pode indicar pequenos ajustes de aprendizado.

2. Acurácia (Accuracy)

O gráfico de acurácia (à direita) mostra que a acurácia de treinamento começa em torno de 0.80 e melhora rapidamente até estabilizar-se em torno de 0.90 após cerca de 30-40 épocas.
A acurácia de validação segue um padrão similar, porém com uma queda acentuada por volta da 30ª época. Isso indica que o modelo teve uma fase em que a generalização não foi tão eficaz, o que pode estar relacionado com as flutuações observadas também na perda.
No entanto, a acurácia de validação recupera-se bem após essa queda e se estabiliza próxima à acurácia de treinamento (~0.92), indicando que, após essas fases iniciais, o modelo consegue generalizar de forma bastante eficiente.

3. Conclusão

O modelo Attention U-Net apresentou um desempenho robusto, com uma queda significativa da perda e aumento na acurácia nas primeiras 40 épocas. As oscilações observadas tanto na perda quanto na acurácia de validação indicam que o modelo passou por algumas fases de dificuldades de generalização, especialmente entre as 30ª e 40ª épocas, mas conseguiu recuperar-se e estabilizar a performance. A estabilização das curvas de perda e acurácia em torno de 0.12-0.15 e 0.90-0.92, respectivamente, sugere que o modelo é capaz de capturar as características do conjunto de dados com eficácia, embora algumas melhorias na consistência da validação possam ser exploradas para reduzir as flutuações observadas.

A utilização de camadas de atenção pode ter contribuído para melhorar o desempenho em identificar detalhes importantes nas imagens, embora ajustes adicionais possam ser necessários para mitigar as oscilações observadas na fase intermediária do treinamento.
